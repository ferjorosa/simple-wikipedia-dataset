```python
from langchain_openai import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from os import getenv
from dotenv import load_dotenv

load_dotenv()

def format_text_to_markdown(model: str, raw_text: str) -> str:
    """
    Converts raw text into clean, properly formatted markdown.

    Args:
        raw_text (str): The input text to be formatted.

    Returns:
        str: The formatted markdown text.
    """
    template = """Transform the following semi-cleaned text into well-formatted Markdown. Follow these instructions carefully:

1. **Titles**: Convert titles indicated with `== Title ==` into Markdown headings. The number of `=` signs corresponds to the heading level. For example:
   - `== Title ==` → `## Title`
   - `=== Subtitle ===` → `### Subtitle`

2. **Lists**: Ensure bullet points (`*`) are properly formatted as Markdown unordered lists.

3. **Code blocks**: Use them if necessary.

4. **Non-English Characters**: Remove any non-English characters (e.g., Chinese, Cyrillic) to simplify the text.

5. **Cleanliness**: Ensure the text is free of unnecessary symbols, extra spaces, or formatting artifacts.

IMPORTANT: Only return the formatted text. Nothing else.

Here’s the text to transform:


    {text}
    """

    prompt = PromptTemplate(template=template, input_variables=["text"])

    llm = ChatOpenAI(
        openai_api_key=getenv("OPENROUTER_API_KEY"),
        openai_api_base="https://openrouter.ai/api/v1",
        model_name=model,
    )

    output_parser = StrOutputParser()

    chain = prompt | llm | output_parser

    return chain.invoke(input={"text": raw_text})
```

```python
import re

def reformat_text(text):

    # Add line breaks before and after titles (1 to 5 levels)
    text = re.sub(r'(={1,5}\s*[^=]+\s*={1,5})', r'\n\n\1\n\n', text)

    # Remove multiple consecutive newlines
    text = re.sub(r'\n{3,}', '\n\n', text)

    # Format lists (e.g., * bullets, # numbers)
    # text = re.sub(r'\n\s*[*#]\s+', '\n* ', text)  # Convert bullets and numbers to a simple bul

    # Strip leading and trailing whitespace
    return text.strip()
```
